```python
def check_duplicates(df, subset=None, show_samples=5):
    # Số duplicate (không tính bản ghi đầu)
    dup_excl_first = df.duplicated(subset=subset, keep='first').sum()
    
    # Số bản ghi thuộc các nhóm trùng (bao gồm cả bản ghi đầu)
    dup_all = df.duplicated(subset=subset, keep=False).sum()

    print(f"Số dòng bị trùng: {dup_excl_first}")
    print(f"Tổng số dòng nằm trong các nhóm trùng: {dup_all}")

    if dup_all > 0:
        duplicates_df = df[df.duplicated(subset=subset, keep=False)]
        
        print("\nMột vài dòng bị trùng:")
        display(duplicates_df.head(show_samples))
        
        return duplicates_df
    else:
        print("Không có dữ liệu trùng.")
        return None
    
check_duplicates(data)
```

```python
def show_unique_and_value_counts(df, top_n=None):
    for col in df.columns:
        if col.lower() == 'id':  # bỏ qua cột id
            continue
        elif col.lower() == 'name' :
            continue
            
        vc = df[col].value_counts(dropna=False)
        unique_count = df[col].nunique(dropna=False)
        print(f"Feature: {col} — Unique values: {unique_count}")
        
        if top_n:
            print(vc.head(top_n).to_string())
        else:
            print(vc.to_string())
        print("-" * 60)

show_unique_and_value_counts(data, top_n=20)
```
